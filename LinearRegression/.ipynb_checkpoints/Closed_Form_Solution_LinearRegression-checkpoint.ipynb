```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples=400, n_features=1, n_informative=1, noise=1.8, random_state=11)
```

```python
Y = Y.reshape((-1,1))
print(X.shape)
print(Y.shape)
```

```python
# Normalising Data
```

```python
X = (X-X.mean())/X.std()
```

```python
# Visualize data
```

```python
plt.figure()
plt.style.use('ggplot')
plt.title('Normalised data')
plt.scatter(X,Y)
plt.show()
```

```python
# appending extra column having one value
ones = np.ones((X.shape[0],1))
X_ = np.hstack((X,ones))
print(X_[:5,:])
```

```python
def hypothesis(X,theta):   # preddiction function
    return np.dot(X,theta)


def getThetaClosedForm(X,Y):
    
    Y = np.mat(Y) # converting Y vector into matrix 
    first = np.dot(X.T,X)
    second = np.dot(X.T,Y)
    
    theta = np.linalg.pinv(first)*second
    # np.(linear_algebra).(pseudo_inverse)
    
    return theta   
```

```python
theta = getThetaClosedForm(X_,Y)
print(theta)  # first is theta1 and second is theta0(bias)
```

```python
plt.figure()
plt.style.use('ggplot')
plt.scatter(X,Y)
plt.plot(X,hypothesis(X_,theta),color='blue')
plt.show()
```
