```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv('dataset/Hardwork_Pays_Off/Training Data/Linear_X_Train.csv').values
Y = pd.read_csv('dataset/Hardwork_Pays_Off/Training Data/Linear_Y_Train.csv').values
```

```python
theta = np.load("ThetaList.npy")
```

```python
plt.ion() # i=interactive mode ; on=on
```

```python
T0 = theta[:,0]
T1 = theta[:,1]
```

```python
for i in range(0,50,3):
    y_ = T0 + T1[i]*X
    
    plt.style.use('ggplot')
    plt.scatter(X,Y)
    
    plt.plot(X,y_,'blue')
    plt.draw()
    plt.pause(1)
    plt.clf() #destroys last object
```
