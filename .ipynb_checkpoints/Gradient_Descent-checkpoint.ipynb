```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2

print(X)
print(Y)
```

**goal** : Given a function f(x), we want to find value of x that minimizes f

### Visualization

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.show()
```

```python
# initialize x 
# step size =  learning rate = lr
```

```python
x = 0
lr = 0.1
error = []  

# 50 steps in the downhill direction
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)  # derivative of y
    x = x-(lr*grad)
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
 # plot the values of error
```

```python
plt.plot(error)
plt.show()
```

```python
# instead of goin for 50 iterations, we can stop early if it reaches a certain threshold
```
