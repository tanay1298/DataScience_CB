### Linear Regression

```python
def hypothesis(x,theta):
    y_ = theta[0] + theta[1]*x
    return y_


def gradient(X,Y,theta):
    m = X.shape[0]
    grad = np.zeros((2,))
    
    for i in range(m):
        x = X[i]
        y_ = hypothesis(x,theta)
        y = Y[i]    
        grad[0] += y_ - y
        grad[1] += (y_ - y)*x
    return grad/m  



def error(X,Y,theta):
    
    total_error = 0.0
    m = X.shape[0]
    
    for i in range(m):
        x = X[i]
        y_ = hypothesis(x,theta)
        y = Y[i]
        total_error += (y_ - y)**2

    return total_error/m

        
    
    
def gradientDescent(X,Y,learning_rate=0.1,max_steps=100):
    
    theta = np.zeros((2,))
    error_list = []
    theta_list = []
    
    for i in range(max_steps):    
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        
        theta_list.append((theta[0],theta[1]))
        error_list.append(e)
            
    return theta,error_list,theta_list



 
```

```python
theta,error_list,theta_list = gradientDescent(X,Y)
```

### Computing Score
Score : R2 (R-Squared) or Coefficient of Determination  

```python
# finding accuracy on training data
```

```python
def r2_score(Y,Y_):
    
    # Instead of loop, n.sum() is recommended as it is fast
    
    num = np.sum( (Y-Y_)**2 )
    deno = np.sum( (Y-Y.mean())**2 )
    
    score = 1 - num/deno
    
    return score
```

```python
score = r2_score(Y,y_)
accuracy = score*100

print(accuracy)
```

### Visualising Loss function, Gradient Descent,, Theta Updates 

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
theta
```

```python
T0 = np.arange(-40,40,1)
T1 = np.arange(40,120,1)
```

```python
# Now we have to generate a surface
```

```python
T0,T1 = np.meshgrid(T0,T1)
```

```python
J = np.zeros(T0.shape)
print(T0.shape)
```

```python
for i in range(J.shape[0]):
    for j in range(J.shape[1]):
        y_ = T0[i,j] + T1[i,j]*X
        J[i,j] = np.sum( (Y-y_)**2 )/Y.shape[0]
    
```

```python
print(J)
```

```python
# Visualize J (loss function)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(T0,T1,J,cmap='rainbow') # Surface Plot
plt.show()
```

```python
# Countour Plot - 
#cut the above bowl shaped surface in parallel planes 
#such that at every level in the plane, 
#the loss fuction has the same value
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(T0,T1,J,cmap='rainbow') # Surface PLot
plt.show()
```
