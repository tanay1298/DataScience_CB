```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Locally Weighted Regression

1) Read and Normalise dataset

2) Generate W for every query point 

3) No training is involvedd, directly make predictions using closed form solution

4) Find the best value of Tau(bandwidth paramater) [Cross Validation]

###  Loading Data 

```python
dfx = pd.read_csv('Dataset/weightedX.csv')
dfy = pd.read_csv('Dataset/weightedY.csv')
```

```python
dfx.head()
```

```python
dfy.head()
```

```python
X = dfx.values
Y = dfy.values
```

```python
print(X.shape) # we have 99 training points each having one feature
print(Y.shape) # we have 99 values for y, each having 1 prediction
```

```python
# Normalising the data
```

```python
u = X.mean()
std = X.std()

X = (X-u)/std
```

```python
X[:5,:]
```

```python
plt.style.use('ggplot')
plt.title('Normalised data')
plt.scatter(X,Y)
plt.show()
```

```python
# thus above data is not linear
```

### Finding W(m x m) diagonal matrix

```python
def getW(query_point,X,tau):
    m = X.shape[0]
    W = np.mat(np.eye(m))
    
    for i in range(m):
        xi = X[i]
        x = query_point
        W[i,i] = np.exp( np.dot((xi-x),(xi-x).T)/(-2*tau*tau) )
        
    return W   
        
```

```python
X = np.mat(X)
Y = np.mat(Y)
m = X.shape[0]

W = getW(-1,X,1)
print(W.shape)
print(W)
```

### Making Predictions 

```python
from scipy import linalg
```

```python
def predict(X,Y,query_x,tau):
    ones = np.ones((m,1))
    X_ = np.hstack([X,ones])
    
    qx = np.mat((query_x,1))
    W = getW(qx,X_,tau)
    
    theta = np.linalg.pinv(X_.T*(W*X_))*(X_.T*(W*Y))    
    pred = np.dot(qx,theta)
    return theta,pred                                          
    
```

```python
theta,pred = predict(X,Y,1.0,1.0)
```

```python
print(theta)
```

```python
print(theta.shape) # theta1, theta0
```

```python
print(pred)
```

### Effect of Bandwidth Parameter 

```python
def plotPrediction(tau):
    X_test = np.linspace(-2,2,20)
    Y_test = []
    
    for xq in X_test:
        theta,pred = predict(X,Y,xq,tau)
        Y_test.append(pred[0][0])
    
    Y_test = np.array(Y_test)
    
    XO = np.array(X)    # X-original
    YO = np.array(Y)
    
    plt.title('Tau/Bandwidth parameter %.2f'%tau)
    plt.scatter(XO,YO)
    plt.scatter(X_test,Y_test,color='blue')
    plt.show()
```

```python
taus = [10,5,1,0.5,0.1]

for t in taus:
    plotPrediction(t)
```

```python
# as the value of tau decreases, wee get a better fit
```
