```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
#Line plot
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color="red",label="apple",linestyle="dashed")
plt.plot(x,y2,color="green",label="kiwi",marker='o')
plt.style.use("seaborn")
plt.xlabel("time")
plt.ylabel("price")
plt.title("prices of fruits over time")
plt.legend()
plt.show()
```

```python
themes = plt.style.available
print(themes)
```

```python
#marker is used to locate the points from which the line is passing
```

```python
prices = np.array([1,2,3,4])**2
print(prices)
```

```python
plt.plot(prices,marker='o')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y1,marker='*',label="apple",color="red")
plt.scatter(x,y2,label="kiwi",color="green")
plt.title("prices over time")
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")
xcor = np.array([0,1,2])*2
plt.bar(xcor-0.25,[10,15,20],width=0.5,label="current year",tick_label=["gold","silver","copper"]) #current year
plt.bar(xcor+0.25,[20,10,30],width=0.5,label="next year") #next year
plt.title("metal comparison")
plt.xlabel("metal")
plt.ylabel("price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn-darkgrid")
subjects = "physics", "chemistry", "maths", "english"
weightage = [10,20,15,25]
plt.pie(weightage,labels=subjects,explode=(0,0,0.5,0),autopct='%1.1f%%',shadow=True)
plt.title("subjeccts")
plt.show()
```
