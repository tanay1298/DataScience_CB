```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read a csv file
```

```python
df = pd.read_csv("dataset/movie_metadata.csv")
```

```python
df.head(10)
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
print(titles[:5])
```

```python
#making a dictioanry to store movie title length and the no of movies of same title length
```

```python
freq_titles = {}

for title in titles:
    length = len(title)
    
    if(freq_titles.get(length)) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
#first row tells that there are 138 movies whose title length is 7 
```

```python
x = np.array(list(freq_titles.keys()))
y = np.array( list(freq_titles.values()) )
```

```python
print(x,y)
```

```python
plt.scatter(x,y)
plt.style.use('ggplot')
plt.xlabel("movie titles length")
plt.ylabel("frequency of titlees length")
plt.title('Movie Visualisation Problem')
plt.show()
```
