```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv('dataset/linearX.csv')
dfy = pd.read_csv('dataset/linearY.csv') 
```

```python
dfx.head()
```

```python
dfy.head()
```

```python
x = dfx.values
y = dfy.values
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
x2 = (x-x.mean())/x.std() #making mean as 0 and vaiance as 1
y2 = (y-y.mean())/y.std()
plt.scatter(x,y,label='raw data')
plt.scatter(x2,y,color='blue',label='normalised data')
plt.legend()
```
